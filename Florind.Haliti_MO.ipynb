## Function 1 - np.vstack 

Add some explanation about the function in your own words
The vstack function stacks arrays horozontally on eachother (row on row) meaning if 2 arrays that are 4 numbers long using vstack would create a 2 dimensional 4 by 2 matrix.

```python
# Example 1 - working 
x = np.array([1, 2, 3, 4])
y = np.array([5, 6, 7, 8])
theStack = np.vstack((x,y))
print(theStack)
theStack.ndim
```

Here we have created to arrays (x and y) and by using the vstack function we hace stacked them on eachother creating a 4 by two array (example above). By using the ndim function we can also see that the object that vstack creates (stacks togeather) is 2 dimensional

```python
# Example 2 - working
x = np.array([[1, 2, 3, 4],[0.1, 0.2, 0.3, 0.4]])
y = np.array([5, 6, 7, 8])
z = np.array([9, 10, 11, 12])
theStack = np.vstack((x,y,z))
print(theStack)
theStack.ndim
```

In the example above we can se that it is possible to have a 2 dim array in a variable. so stacking a 2 dimensional array with one dimensional arrays as the results printed is still a 2 dim array.  

```python
# Example 3 - breaking (to illustrate when it breaks)
x = np.array([[[1, 2, 3, 4], [21,22,23,24]] ,[[0.1, 0.2, 0.3, 0.4],[31,32,33,34]]])
y = np.array([[5, 6, 7, 8], [9, 10, 11, 12]])
theStack = np.vstack((x,y))
print(theStack)
theStack.ndim
```

Explanation about example (why it breaks and how to fix it)
The error occurs because a 3 dimensional array is vertically stacked on a 2 dimensional array. The Vstack function is not able to stack two arrays with different dimensions. This problem can be fixed by converting the arrays so they have the same number of dimensions, either converting y to a 3 dim array or x to a 2 dim array. another error can occur when the number of columns in one array is not similar to the other example, stacking a 1 by 3 array with a 1 by 4, however this is not relevant to this error but worth mentioning.
